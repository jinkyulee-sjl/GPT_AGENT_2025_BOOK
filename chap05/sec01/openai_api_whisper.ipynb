```python
# from openai import OpenAI
from google import genai
from dotenv import load_dotenv
import os

load_dotenv()
# api_key = os.getenv("OPENAI_API_KEY")
# client = OpenAI(api_key=api_key)

try:
    client = genai.Client()
except Exception as e: 
    print("Error initializing Google GenAI client:", e)
    exit(1)

def generate_gemini_response(prompt: str):
    # 사용할 모델 지정
    model = 'gemini-2.5-flash'
    
    # 모델에 콘텐츠 생성 요청
    response = client.models.generate_content(
        model=model,
        contents=[prompt]
    )
    
    return response.text

# 4. 사용 예시
user_prompt = "간단한 인삿말을 한국어로 해줘."
try:
    result = generate_gemini_response(user_prompt)
    print(f"--- Gemini Response (Model: gemini-2.5-flash) ---")
    print(result)

except Exception as api_error:
    # API 호출 중 네트워크, 인증, 요금 등의 문제가 발생했을 때 오류를 출력
    print(f"An API error occurred: {api_error}")
```
